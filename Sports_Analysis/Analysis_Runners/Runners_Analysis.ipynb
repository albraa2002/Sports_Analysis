To find approximate exercise time, I will filter for intensity level of at least 2 (fairly active), see if users have a habit of workout at certain time. I choose not to start with level 1 (lightly active). Though it could make more sense at the beginning at the workout, this level of intensity could happen with daily activities other than workout as well.

```python
times = workout[['Id', 'dt-datetime']]
times
```

We will create a new data frame and keep track start and end time of moderate and higher intensity (2 and up). This new data frame is named tracked_workout, and includes Id, date, start time, end time. We will then derive workout duration, and number of workouts a day based on tracked_workout data frame.

```python
wko_start_times = []
wko_end_times = []
wko_ids = []
wko_dates = []
```

```python
wko_start_times = []
wko_end_times = []
wko_ids = []
wko_dates = []
user = Ids[0]
one_user = times.loc[times['Id'] == user][['Id', 'dt-datetime']]
print('Len(first_user): {}'.format(len(one_user)))
start_time = one_user.iloc[0]['dt-datetime']
print('First start time: {}'.format(start_time))
wko_start_times.append(start_time)
wko_ids.append(user)
wko_dates.append(str(start_time)[:10])
for i in list(range(1, len(one_user))):
    t = one_user.iloc[i]['dt-datetime']
    delta = t - one_user.iloc[i-1]['dt-datetime']
    if (delta.seconds >= 1800):
        end_time = one_user.iloc[i-1]['dt-datetime']
        wko_end_times.append(end_time)
        start_time = t
        wko_start_times.append(start_time)
        print('Delta: {}'.format(delta))
        print('End time: {} Start time: {}'.format(end_time, start_time))
        wko_ids.append(user)
        wko_dates.append(str(start_time)[:10])
   # if (wko_end_times[-1] != one_user.iloc[-1]['dt-datetime']):
wko_end_times.append(one_user.iloc[-1]['dt-datetime'])
print('Last end time: {}'.format(one_user.iloc[-1]['dt-datetime']))
print('User: {} len(wko_start_times): {} len(wko_end_times): {}'.format(user, len(wko_start_times), len(wko_end_times)))
```

```python
wkotime_cleaned = workout_tracktime.loc[workout_tracktime['duration'] < pd.Timedelta(hours=2)]
wkotime_cleaned.describe()
```

```python
sns.histplot(data=wkotime_cleaned, x='duration(m)', binwidth=5).set(title='Workout duration distribution')
```

10-15 minute duration is the most popular, with over 300 workout sessions. The second most popular duration is 15-20 minutes, with around 140 sessions. The third most popular duration is 30-35 minutes, with around 100 sessions.
Our statistics also shows that:
50% of workout sessions last 25 minutes or less.
75% of workout sessions last 45 minutes or less.
Based on these findings, we recommend Bella Beats create workout sessions with below duration:
- Short workout: 10 and 15 minutes
- Medium workout: 20 minutes
- Advanced workout: 30 and 40 minutes

We should focus the most on the short workouts, then the medium workouts. Advanced workout with longer duration is nice to have for above-average users, but needs not to be the heaviest investment.
